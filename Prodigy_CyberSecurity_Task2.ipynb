```python
#Pixel Manipulation for Image Encryption
#Requisites:
#Perform any mathematical operation for all pixels.
#Allow users to encrypt and decrypt the images.
```

```python
import numpy as np
def encrypt_image(image_path, key):
    img = Image.open(image_path)
    data = np.array(img)
    
    for i in range(512):
        for j in range(512):
            data[i][j]+=key
            data[i][j]+=i
    data=np.rot90(data,2)
    encrypted_image = Image.fromarray(data)
    encrypted_image.save('encrypted_image.png')
    return encrypted_image

def decrypt_image(key):
    img = Image.open('encrypted_image.png')
    data = np.array(img)
    data=np.rot90(data,2)
    for i in range(512):
        for j in range(512):
            data[i][j]-=key
            data[i][j]-=i
    
    decrypted_image = Image.fromarray(data)
    decrypted_image.save('decrypted_image.png')
    return decrypted_image

image_path = 'lena.png'
key = 500000


r=encrypt_image(image_path, key)

d=decrypt_image(key)
display(r)
display(d)
```
